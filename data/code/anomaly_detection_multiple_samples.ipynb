```python
import os
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 读取 Hex.xlsx 文件
data1 = pd.read_excel('Hex.xlsx')

# 创建 Isolation Forest 模型
clf = IsolationForest(contamination=0.05, random_state=42)

# 创建一个空的列表来存储所有文件的异常点索引
all_indices = []

# 设置拆分数据所在的文件夹路径
input_folder = 'Split Data'

# 循环遍历拆分数据文件夹中的每个文件
for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)

    # 读取拆分数据文件
    new_data = pd.read_excel(file_path)

    # 将数据中的非数字值转换为 NaN
    new_data = new_data.apply(pd.to_numeric, errors='coerce')

    # 合并数据
    combined_data = pd.concat([data1.iloc[:, :30], new_data.iloc[:, :30]], ignore_index=True)

    # 使用 PCA 将数据降维到 3 维
    pca = PCA(n_components=3)
    features_3d = pca.fit_transform(combined_data)

    # 对数据进行 Z-score 标准化
    scaler = StandardScaler()
    z_scores = scaler.fit_transform(features_3d)
    
    z_outliers = np.where((np.abs(z_scores) > 3).any(axis=1))


    # 如果有异常点，则将其索引添加到列表中
    if len(z_outliers) > 0:
        for index in z_outliers:
            all_indices.append({'Filename': filename, 'Outlier Index': index})

# 将所有异常点索引转换为 DataFrame
indices_df = pd.DataFrame(all_indices)

# 将异常点索引保存到 Excel 文件中的 Sheet1
output_excel_path = 'Z_Score_Outlier_Indices.xlsx'
with pd.ExcelWriter(output_excel_path) as writer:
    indices_df.to_excel(writer, sheet_name='Sheet1', index=False)

print(f"Z-score outlier indices saved to '{output_excel_path}'")
```

```python
import os
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import LocalOutlierFactor

# 读取 Hex.xlsx 文件
data1 = pd.read_excel('Hex.xlsx')

# 创建一个空的列表来存储所有文件的异常点索引
all_indices = []

# 设置拆分数据所在的文件夹路径
input_folder = 'Split Data'

# 循环遍历拆分数据文件夹中的每个文件
for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)

    # 读取拆分数据文件
    new_data = pd.read_excel(file_path)

    # 将数据中的非数字值转换为 NaN
    new_data = new_data.apply(pd.to_numeric, errors='coerce')

    # 合并数据
    combined_data = pd.concat([data1.iloc[:, :30], new_data.iloc[:, :30]], ignore_index=True)

    # 使用 PCA 将数据降维到 3 维
    pca = PCA(n_components=3)
    features_3d = pca.fit_transform(combined_data)



    # 创建 LOF 模型
    lof = LocalOutlierFactor(n_neighbors=10, contamination=0.05)

    # 计算 LOF 分数
    lof_outlier_scores = lof.fit_predict(features_3d)

    # 获取 LOF 异常点的索引
    lof_outliers = np.where(lof_outlier_scores == -1)[0]

    # 如果有异常点，则将其索引添加到列表中
    if len(lof_outliers) > 0:
        for index in lof_outliers:
            all_indices.append({'Filename': filename, 'Outlier Index': index})

# 将所有异常点索引转换为 DataFrame
indices_df = pd.DataFrame(all_indices)

# 将异常点索引保存到 Excel 文件中的 Sheet1
output_excel_path = 'LOF_Outlier_Indices.xlsx'
with pd.ExcelWriter(output_excel_path) as writer:
    indices_df.to_excel(writer, sheet_name='Sheet1', index=False)

print(f"LOF outlier indices saved to '{output_excel_path}'")
```

```python
import os
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

# 读取 Hex.xlsx 文件
data1 = pd.read_excel('Hex.xlsx')

# 创建一个空的列表来存储所有文件的异常点索引
all_indices = []

# 设置拆分数据所在的文件夹路径
input_folder = 'Split Data'

# 循环遍历拆分数据文件夹中的每个文件
for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)

    # 读取拆分数据文件
    new_data = pd.read_excel(file_path)

    # 将数据中的非数字值转换为 NaN
    new_data = new_data.apply(pd.to_numeric, errors='coerce')

    # 合并数据
    combined_data = pd.concat([data1.iloc[:, :30], new_data.iloc[:, :30]], ignore_index=True)

    # 使用 PCA 将数据降维到 3 维
    pca = PCA(n_components=3)
    features_3d = pca.fit_transform(combined_data)



    # 创建 KNN 模型
    knn = NearestNeighbors(n_neighbors=10)
    knn.fit(features_3d)

    # 计算 KNN 分数
    knn_outlier_scores = knn.kneighbors(features_3d)[0][:, -1]

    # 计算异常点阈值
    knn_threshold = np.percentile(knn_outlier_scores, 95)

    # 获取 KNN 异常点的索引
    knn_outliers = np.where(knn_outlier_scores > knn_threshold)[0]

    # 如果有异常点，则将其索引添加到列表中
    if len(knn_outliers) > 0:
        for index in knn_outliers:
            all_indices.append({'Filename': filename, 'Outlier Index': index})

# 将所有异常点索引转换为 DataFrame
indices_df = pd.DataFrame(all_indices)

# 将异常点索引保存到 Excel 文件中的 Sheet1
output_excel_path = 'KNN_Outlier_Indices.xlsx'
with pd.ExcelWriter(output_excel_path) as writer:
    indices_df.to_excel(writer, sheet_name='Sheet1', index=False)

print(f"KNN outlier indices saved to '{output_excel_path}'")
```
